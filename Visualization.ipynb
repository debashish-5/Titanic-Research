```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('updates_titanic.csv')
```

```python
df
```

### Starting visualization

#### 1-Histplot(Visualixation between Gender And Ticket Price

```python
x = df.groupby('Sex')['Fare'].mean()
```

```python
x
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(12,6))
sns.histplot(x=x.index,y = x.values,color='blue')
plt.title('Ticket Fare Vs Gender')
plt.xlabel('Gender')
plt.ylabel('Ticket Fare')
```

```python
Alive = df[df['Survived']==1]
```

```python
dead = df[df['Survived']==0]
```

```python
y1 = dead['Survived']
```

```python
y2 = dead['Age']
```

```python
y_label = dead['Survived'].count()
```

```python
y_label
```

```python
plt.figure(figsize=(20,8))
plt.grid(True)
sns.barplot(x = x1,y = x2,label='123 Person Alive ')
sns.barplot(x=y1,y = y2,label='60 Person Dead')
plt.title("Visual Between Alive Person And Dead Person's Age")
plt.tight_layout()
plt.legend()
```
